````java
//Escriu la teva classe Propulsor aquí

class Propulsor {
    private int potencia; //potencia actual que té en cada moment
    
    private boolean arrencat;
    
    //potència màxima que pot tenir el propulsor
    //És final perquè un cop creat l'objecte, no es pot modificar
    private final int potenciaMaxima; 
    
    //Constructor
    public Propulsor(int potenciaMaxima){
        this.potenciaMaxima = potenciaMaxima;
    }
    
    //setters
    public void setArrencat(boolean arrencat){
        this.arrencat = arrencat;
    }
    
    //(potenciaMaxima no té setter perquè és final)
    public void setPotencia(int potencia){
        this.potencia = potencia;
    }
    
    //getters
    public boolean getArrencat(){
        return this.arrencat;
    }
    
    public int getPotencia(){
        return this.potencia;
    }
    
    public int getPotenciaMaxima(){
        return this.potenciaMaxima;
    }
}
````

````java
//Escriu la teva classe Coet aquí
//Comença pels canvis de la fase 1

class Coet {
    //El teu codi aquí
    //Atributs:
    private final String id; //és immutable
    
    //Ha de ser un array de mida 3
    //Inicialment no conté cap propulsor
    //(inicialment conté valors null)
    private final Propulsor[] propulsors = new Propulsor[3]; 
    
    //Constructor:
    public Coet(String id){
        this.id = id;
    }
    
    public void arrenca(int potencia0, int potencia1, int potencia2) {        
        
        //Primer comprovem que tenim tots tres propulsors
        for(Propulsor propulsor : this.propulsors) {
            if(propulsor == null) {//Si és null, encara no s'ha assignat un propulsor
                System.out.println("Error: aquest coet encara no té propulsors i no pot ser arrencat.");
                return; //acaba i no executa la resta
            }
        }
        
        //Segon, comprovem que les potències passades són vàlides
        //(la potència és vàlida si no és negativa i no és superior a la màxima del propulsor)
        if(potencia0 < 0 ||
           potencia1 < 0 ||
           potencia2 < 0 ||
           potencia0 > this.propulsors[0].getPotenciaMaxima() ||
           potencia1 > this.propulsors[1].getPotenciaMaxima() ||
           potencia2 > this.propulsors[2].getPotenciaMaxima()) {
                System.out.println("Error: potència assignada a propulsor no vàlida.");
                return; //acaba i no executa la resta
        }        
        
        //Si arribem aquí vol dir que tots 3 propulsors han estat inicialitzats correctament
        //i que les potències són vàlides
        //(la inicialització dels propulsors es fa amb el setter dels propulsors)
        this.propulsors[0].setPotencia(potencia0);
        this.propulsors[0].setArrencat(true);
        
        this.propulsors[1].setPotencia(potencia1);
        this.propulsors[1].setArrencat(true);
        
        this.propulsors[2].setPotencia(potencia2);
        this.propulsors[2].setArrencat(true);
    }
    
    public void atura() {
        this.propulsors[0].setPotencia(0);
        this.propulsors[0].setArrencat(false);
        
        this.propulsors[1].setPotencia(0);
        this.propulsors[1].setArrencat(false);
        
        this.propulsors[2].setPotencia(0);
        this.propulsors[2].setArrencat(false);
    }
    
    public void printStatus() {
        
        //Càlcul de la suma de potències de tots els propulsors
        //Càlcul de la suma de les potències màximes de tots els propulsors
        int sumaPotencies = 0;
        int sumaPotenciesMaximes = 0;
        for(Propulsor propulsor : this.propulsors){
            sumaPotencies += propulsor.getPotencia();
            sumaPotenciesMaximes += propulsor.getPotenciaMaxima();
        }
        
        System.out.println("Coet amb id " + id + 
                           " en marxa: " + this.propulsors[0].getArrencat() + 
                           ", potència " + sumaPotencies + 
                           ", potència màxima " + sumaPotenciesMaximes);
    }
    
    //setter
    public void setPropulsors(Propulsor[] propulsors){
        for(int i = 0; i < 3; i++) //Assignem un a un els tres propulsors
            this.propulsors[i] = propulsors[i];
    }
    
    //getters
    public String getId(){
        return this.id;
    }
}
````

````java
//Escriu les teves proves aquí

Coet coet = new Coet("32WESSDS");

Propulsor propulsor1 = new Propulsor(50);
Propulsor propulsor2 = new Propulsor(30);
Propulsor propulsor3 = new Propulsor(40);

coet.arrenca(20, 35, 30);
//"Error: aquest coet encara no té propulsors i no pot ser arrencat."
    
//Assignem els propulsors al coet amb el setter
coet.setPropulsors(new Propulsor[]{propulsor1, propulsor2, propulsor3});


coet.arrenca(20, 35, 30);
//"Error: potència assignada a propulsor no vàlida."
    
coet.arrenca(20, 15, 30); 
coet.printStatus();
//"Coet amb id 32WESSDS en marxa: true, potència 65, potència màxima 120"
    
coet.atura();
//"Coet amb id 32WESSDS en marxa: false, potència 0, potència màxima 120"
````
